### Exercise 1 

Write a function called nested_sum that takes a list of lists of integers and adds up the elements from all of the nested lists.

```python
def nested_sum(list):
    total = 0
    for item in list:
        for i in item:
            total += i
    return total
```

```python
t = [[1, 2], [3], [4, 5, 6]]
nested_sum(t)
```

### Exercise 2

Write a function called cumsum that takes a list of numbers and returns the cumulative sum; that is, a new list where the ith element is the sum of the first i+1 elements from the original list.

```python
def cumsum(list):
    total = 0
    list2 = []
    for i in list:
        total += i
        list2.append(total)
    return list2
```

```python
t = [1, 2, 3]
cumsum(t)
```

### Exercise 7

Write a function called has_duplicates that takes a list and returns True if there is any element that appears more than once. It should not modify the original list.

```python
def has_duplicates(list):
    list_copy = list
    list_copy.sort()

    for i in range(len(list_copy) - 1):
        if list_copy[i] == list_copy[i + 1]:
            return True
    return False
```

```python
t = [1, 2, 3, 4, 4]
has_duplicates(t)
```

```python
t = [1, 2, 3, 4, 5]
has_duplicates(t)
```

If there are 23 students in your class, what are the chances that two of them have the same birthday? You can estimate this probability by generating random samples of 23 birthdays and checking for matches.

Hint: you can generate random birthdays with the randint function in the random module.

```python
from random import randint

def birthdays(n):
    bds = []
    for day in range(n):
        bds.append(randint(1, 365))
    return sorted(bds)
```

```python
def birthday_paradox(samples, universe):
    matches = 0
    for i in range(samples):
        sample = birthdays(universe)
        if has_duplicates(sample):
            matches += 1
    return print(f'{matches} of {samples} simulations found a match')
```

```python
birthday_paradox(1000, 23)
```

### Exercise 9

Write a function that reads the file words.txt and builds a list with one element per word. Write two versions of this function, one using the append method and the other using the idiom t = t + [x]. Which one takes longer to run? Why?

```python
import time
```

```python
def wordlist1():
	wordlist = []
	file = open('words.txt')
	for line in file:
		word = line.strip()
		wordlist.append(word)
	return wordlist

start_time = time.time()
t = wordlist1()
elapsed_time = time.time() - start_time

print(len(t))
print(t[:10])
print(elapsed_time, 'seconds')
```

```python
def wordlist2():
	wordlist = []
	file = open('words.txt')
	for line in file:
		word = line.strip()
		wordlist = wordlist + [word]
	return wordlist

start_time = time.time()
t = wordlist2()
elapsed_time = time.time() - start_time

print(len(t))
print(t[:10])
print(elapsed_time, 'seconds')
```

### Exercise 10

To check whether a word is in the word list, you could use the in operator, but it would be slow because it searches through the words in order.

Because the words are in alphabetical order, we can speed things up with a bisection search (also known as binary search), which is similar to what you do when you look a word up in the dictionary (the book, not the data structure). You start in the middle and check to see whether the word you are looking for comes before the word in the middle of the list. If so, you search the first half of the list the same way. Otherwise you search the second half.

Either way, you cut the remaining search space in half. If the word list has 113,809 words, it will take about 17 steps to find the word or conclude that it’s not there.

Write a function called in_bisect that takes a sorted list and a target value and returns True if the word is in the list and False if it’s not.

Or you could read the documentation of the bisect module and use that!

```python
def make_wordlist():
	wordlist = []
	file = open('words.txt')
	for line in file:
		word = line.strip()
		wordlist.append(word)
	return wordlist

def binary_search(wordlist, word):
    if len(wordlist) == 0:
        return False

    i = len(wordlist) // 2
    if wordlist[i] == word:
        return True

    if wordlist[i] > word:
        return binary_search(wordlist[:i], word)                                      
    else:
        return binary_search(wordlist[i+1:], word)
```

```python
wordlist = make_wordlist()
binary_search(wordlist, 'aa')
```

### Exercise 11

Two words are a “reverse pair” if each is the reverse of the other. Write a program that finds all the reverse pairs in the word list.

```python
def reverse_pair(wordlist, word):
    return binary_search(wordlist, word[::-1])
```

### Exercise 12

Two words “interlock” if taking alternating letters from each forms a new word. For example, “shoe” and “cold” interlock to form “schooled”.

Write a program that finds all pairs of words that interlock. Hint: don’t enumerate all pairs!

```python
def interlock(wordlist, word):
    return binary_search(wordlist, word[::2]) and binary_search(wordlist, word[1::2])
```

```python
for word in wordlist:
    if interlock(wordlist, word):
        print(word, word[::2], word[1::2])
```
