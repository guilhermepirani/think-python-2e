### 2) Suppose the cover price of a book is 24.95, but bookstores get a 40 percent discount. Shipping costs 3 for the first copy and 75 cents for each additional copy. What is the total wholesale cost for 60 copies?

```
cover_price = 24.95
discount = 40/100
shipping = 3
shipping_extra = 0.75  # for copy
order = 60

print("Price for Bookstores:", (cover_price - (cover_price * discount)) *
      order + shipping + (shipping_extra * (order - 1)))
print("Price for other buyers:", (cover_price * order + shipping + 
                                  (shipping_extra * (order - 1))))
```
